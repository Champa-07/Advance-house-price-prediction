## House prices : Advanced Regression Techniques 

```python
##Data Analysis phase 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns 

##display all the columns of dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('house prediction train.csv.csv')
```

```python
print(dataset.shape)
```

```python
dataset.head()
```

### Missing Values

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for features in features_with_na:
    print(features,np.round(dataset[features].isnull().mean(),4), '% missing values')
```

### Lets plot a graph and find a relationship between missing values and sales price

```python
for features in features_with_na:
    data=dataset.copy()
    
    data[features] = np.where(data[features].isnull(),1,0)
    
    data.groupby(features)['SalePrice'].median().plot.bar()
plt.title(features)
plt.show()
```

### Numerical Variables
                           
                        

```python
numerical_features = [features for features in dataset.columns if dataset[features].dtypes !='O']## caps o 

print('Number of numerical variables: ',len(numerical_features))

dataset[numerical_features].head()
```

```python
## List of variables that contain year information 
year_feature = [features for features in numerical_features if 'Yr' in features or 'Year' in features]

year_feature
```

```python
for features in year_feature:
    print(features,dataset[features].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House price')
plt.title('House price vs Year Sold')
```

```python
for features in year_feature:
 if features!= 'YrSold':
        data=dataset.copy()
        data[features]=data['YrSold']-data[features]
        
        plt.scatter(data[features],data['SalePrice'])
        plt.show()
```

```python
discrete_feature=[features for features in numerical_features if len(dataset[features].unique())<25 and features not in year_feature+['Id']]

discrete_feature
```

```python
for features in discrete_feature:
    data=dataset.copy()
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.title(features)
    plt.show()
```

```python
continous_feature= [features for features in numerical_features if features not in discrete_feature+year_feature+['id']]
continous_feature
```

```python
for features in continous_feature:
    data=dataset.copy()
    data[features].hist(bins=25)
    plt.ylabel("Count")
    plt.xlabel(features)
    plt.title(features)
    plt.show()
```
